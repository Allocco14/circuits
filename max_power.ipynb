## Gráfica de potencia máxima

```python
import numpy as np
import matplotlib.pyplot as plt

## Partiendo de un circuito Thévenin con una resistencia de 100 ohmios y una fuente de 10V
## Se desea calcular la potencia por una potencia en el rango 1 a 100 % de R1 (100 ohmios)
## para así obtener el valor de la potencia máxima

V = 10
r1 = 100

## Arreglo de 1 a 100
r2 = np.arange(0, 1001, 1)

P = (V**2 * r2) / (r1 + r2)**2

print("La potencia máxima es: ", P.max())

 ## Gráfica

plt.plot(r2, P)
plt.xlabel('R2 (ohmios)')
plt.ylabel('Potencia (W)')
plt.title('Potencia en función de R2')
plt.grid()
```
